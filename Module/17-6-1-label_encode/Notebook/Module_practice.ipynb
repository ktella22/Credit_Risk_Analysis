```python
# Import dependencies
import pandas as pd
from path import Path

file_path = Path("../Resources/loans_data.csv")
loan_df = pd.read_csv(file_path)
loan_df.head()
```

```python
# DataFrame size
loan_df.shape
```

```python
# convert textual into numerical values on Gender column 
loans_binary_encoded = pd.get_dummies(loan_df, columns=["gender"])
loans_binary_encoded.head()
```

```python
# Encoding on education column using get_dummbies() module
loans_binary_encoded = pd.get_dummies(loan_df, columns=["education", "gender"])
loans_binary_encoded.head()
```

```python
# Using LabelEncoder to transform text into numberical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = loan_df.copy()
df2["education"] = le.fit_transform(df2["education"])
df2.head()
```

```python
# To encode gender column 
df2["gender"] = le.fit_transform(df2["gender"])
df2.head()
```

```python
# Use laberencoder on loan_df 
label_encoder = LabelEncoder()
loan_df["month_le"] = label_encoder.fit_transform(loan_df["month"])
loan_df.head()
```

```python
# customize encoding to the "Month" column 
# Create dictionary and apply that to the column that wanna be encoded
months_num ={"January": 1, 
            "February": 2,
             "March": 3, 
             "April": 4, 
             "May": 5, 
             "June": 6, 
             "July": 7,
             "August": 8, 
             "September": 9, 
             "October": 10, 
             "November": 11, 
             "December": 12
            }
```

```python
# Call lambda function 
loan_df["month_num"] = loan_df["month"].apply(lambda x: months_num[x]) 
loan_df.head()
```

```python
# Drop the column "month" 
loan_df = loan_df.drop(["month", "month_le"], axis=1)
loan_df.head()
```
